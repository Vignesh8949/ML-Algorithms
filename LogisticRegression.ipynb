```python
import pandas as pd
import numpy as np
dataset = pd.read_csv('User_Data.csv')
dataset
```

```python
dataset['Gender'] = dataset['Gender'].map({'Male': 0,'Female': 1, }).astype(float)
dataset
```

```python
x = dataset.iloc[:,[2,3]].values
y = dataset.iloc[:,4].values
```

```python
y
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.25)
```

```python
ytrain
```

```python
xtrain
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
xtrain = sc_x.fit_transform(xtrain)
xtest = sc_x.transform(xtest)

print(xtrain[0:10, :])
print(xtest[0:10, :])
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(xtrain,ytrain)
```

```python
import pickle
with open('Logistic_model.pk1', 'wb')as files:
    pickle.dump(classifier,files)
with open('Logistic_model.pk1', 'rb')as files:
    classifier = pickle.load(files)
```

```python
y_pred = classifier.predict(xtest)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest, y_pred)
print('confusion matrix : \n', cm)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(ytest, y_pred))
```

```python
from sklearn.metrics import precision_score
print("precion:", precision_score(ytest, ytest))
```

```python
from sklearn.metrics import recall_score
print("Recallscore:", recall_score(ytest, ytest))
```
